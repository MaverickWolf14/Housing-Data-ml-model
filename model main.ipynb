## Split data

now let's split the data into two sets one train set(80%) nad test set(20%)
 we will use stratified spliting and use CHAS as refrence attribute 
 Also we will need to stratified the random function while splitting so that data in both set don't change when we re run the code

```python
from sklearn.model_selection import StratifiedShuffleSplit

split= StratifiedShuffleSplit(n_splits=1, test_size = 0.2, random_state = 42)
# her we said that we willl split in once in ratio 0.2 and random state is 42 so that the data don't change after spliting

for train_index, test_index in split.split(housing_data,housing_data['CHAS']):
    strat_train_set= housing_data.loc[train_index]
    strat_test_set = housing_data.loc[test_index]
```

## What we did

## Fixing the missing data

we observed there is missing data in RM data attribute 

## there are three ways we can fix missing data

> 1. Deleting the data point : we can do that when very less no. of data points are missing

> 2. Deleting the attribute :  we can do this when the attribute is not that important for the data se and for the prtediction

> 3. Filling the missing data points:  we can do this by filling mean median or any reasonable data in the missing data set

## Solution 

WE will use 3rd optio and will fill median in the missing data point

we can do this in two ways we can just enter the media in missing data points of RM attribute 

or

we can make pipeline which will fill all the missing data points from all the attributes with there median

## we will make the pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])
```

```python
housing_data_tr= my_pipeline.fit_transform(housing_data)
housing_data_tr
```

The data is now in array format

## Selectig a model that will suit our data

we will try 

1. Linear Regression 
2. DecissionTree regression
3. Random Forest Tree

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

model1 = LinearRegression()
model1.fit(housing_data_tr,housing_data_label)

model2 = DecisionTreeRegressor()
model2.fit(housing_data_tr, housing_data_label)

model3 = RandomForestRegressor()
model3.fit(housing_data_tr,housing_data_label)
```

## Here we added three models and will check which one gives us more accuracy and willb use that one as our finall model

## now we will take a small data set and will these models to predict the data|

```python
small_data= housing_data.iloc[:5]
small_label = housing_data_label.iloc[:5]
prepared_data = my_pipeline.transform(small_data)

#notice we called the data from housing_data not housing_data_tr
# hence there are chances that we can call data where data from RM attribute is missig
# to avoid that we are trasforming the data by passing it through the pipeline
```

## Now we will evaluate the models and check the chances of errors

```python
from sklearn.metrics import mean_squared_error
prediction1=model1.predict(housing_data_tr)
lin_mse1=mean_squared_error(housing_data_label, prediction1)
lin_rmse1 = np.sqrt(lin_mse1)
lin_rmse1
```

```python
prediction2 = model2.predict(housing_data_tr)
lin_mse2 = mean_squared_error(housing_data_label, prediction2)
lin_rmse2 = np.sqrt(lin_mse2)
lin_rmse2
```

```python
prediction3 = model3.predict(housing_data_tr)
lin_mse3 = mean_squared_error(housing_data_label, prediction3)
lin_rmse3 = np.sqrt(lin_mse3)
lin_rmse3
```

## Here we can see that 

error in model 1 is bit high 

error in model 2 is 0 which shows the model had memorized the data as there were correlation and there should be some error in the model if it understand the trend

error in model 3 is relatively low

## Using crossvalidaion for checking the error

```python
from sklearn.model_selection import cross_val_score
score1 = cross_val_score(model1, housing_data_tr, housing_data_label, scoring="neg_mean_squared_error", cv=10)
score2 = cross_val_score(model2, housing_data_tr,
                         housing_data_label, scoring="neg_mean_squared_error", cv=10)
score3 = cross_val_score(model3, housing_data_tr,
                         housing_data_label, scoring="neg_mean_squared_error", cv=10)
```

```python
def print_score(Scores):
    rmse_scores = np.sqrt(-Scores)
    print("Score: ",rmse_scores)
    print("Mean: ", rmse_scores.mean())
    print("Standard Deviation: ", rmse_scores.std())
```

## Printing scores : model 1 

```python
print_score(score1)
```

## Printing scores : model 2

```python
print_score(score2)
```

## Printing scores : model 3

```python
print_score(score3)
```
